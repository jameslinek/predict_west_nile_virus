**Now, let's add a Date Separator function (Courtesy of Brian Collins) to pull the years/months/days of each date into separate columns.**

```python
def date_separate(df):
    df = df.copy()
    df['Year'] = pd.DatetimeIndex(df['Date']).year
    df['Month'] = pd.DatetimeIndex(df['Date']).month
    df['Day'] = pd.DatetimeIndex(df['Date']).day
    return df
```

```python
# Apply the entire dataset to this new function.

train=date_separate(train)
test=date_separate(test)
weather=date_separate(weather)
spray=date_separate(spray)
```

### Now, let's merge the Weather data onto the Train and Test data horizontally:

```python
train = train.merge(weather, on='Date')
test = test.merge(weather, on='Date')
train = train.drop(['Date'], axis = 1)
test = test.drop(['Date'], axis = 1)
```

```python
# Using a for loop again, let's create a function that defines mosquitos with West Nile and those without:

def bug_find(df):
    for bug in df:
        if bug == 'CULEX PIPIENS/RESTUANS':
            return 'wnv_bug'
        elif bug == 'CULEX PIPIENS':
            return 'wnv_bug'
        elif bug == 'CULEX RESTUANS':
            return 'wnv_bug'
        elif bug == 'UNSPECIFIED CULEX':
            return 'wnv_bug'
        else:
            return 'no_wnv_bug'

train['Species'] = train[['Species']].apply(bug_find, axis=1)
test['Species'] = test[['Species']].apply(bug_find, axis=1)
print(train['Species'].value_counts())
print(test['Species'].value_counts())
```
